#colab 사용법

* MAC을 사용하는 경우 Ctrl 대신 Command입니다. Alt 대신 Option입니다.
1. **실행 단축키**
    1. (필수) `Ctrl(Command)` + `Enter` : 해당 셀 실행
    2. `Shift` + `Enter` : 해당 셀 실행 + 커서를 다음 셀로 이동
    3. (필수) `Alt(Option)` + `Enter` : 해당 셀 실행 + 코드 불록 하단 추가
2. **셀 삽입/삭제 관련 단축키**
    1. `Ctrl(Command)` + M A : 코드 셀 위에 삽입
    2. `Ctrl(Command)` + M B : 코드 셀 아래 삽입
    3. `Ctrl(Command)` + M D : 셀 지우기
    4. `Ctrl(Command)` + M Y : 코드 셀로 변경
    5. `Ctrl(Command)` + M M : 마크다운 셀로 변경
    6. `Ctrl(Command)` + M Z : 실행 취소
3. 수정 관련된 단축키
    1. `Ctrl(Command)` + `Alt(Option)` + 화살표위아래 : 동시 수정
    2. (자주) `Ctrl(Command)` + D : 같은 단어 찾아 동시 수정
    3. `Ctrl(Command)` + `Shift` + L : 동일 단어를 전체로 찾아 동시 수정
    4. `Alt(Option)` + `Shift` + 화살표위아래 : 해당내용을 위나 아래 복사해서 붙여넣기
    5. `Alt(Option)` + 화살표위아래 : 해당 내용을 위나 아래로 보내기
    6. (자주) `Ctrl(Command)` + `Alt(Option)` + 화살표위아래 : 위아래 동시 수정
    7. (자주) Home, End : 문장의 양 끝
    8. (필수) `Ctrl` + `/` : 주석
    9. (필수) `Shift` + `Del` : 한 줄 지우기
    10. (필수) `Tab`, `Ctrl` + `]` : 들여쓰기
    11. (필수) `Shift` + `Tab`, `Ctrl` + `[` : 내어쓰기
4. 단축키 보기 및 설정
    1. `Ctrl(Command)` + M H : 단축키 모음

```python
def connect_to_next_port(self, minimum: int) -> int:
    """Connects to the next available port.

    Args:
      minimum: A port value greater or equal to 1024.

    Returns:
      The new minimum port.

    Raises:
      ConnectionError: If no available port is found.
    """
    if minimum < 1024:
      # Note that this raising of ValueError is not mentioned in the doc
      # string's "Raises:" section because it is not appropriate to
      # guarantee this specific behavioral reaction to API misuse.
      raise ValueError(f'Min. port must be at least 1024, not {minimum}.')
    port = self._find_next_open_port(minimum)
    if port is None:
      raise ConnectionError(
          f'Could not connect to service on port {minimum} or higher.')
    assert port >= minimum, (
        f'Unexpected port {port} when minimum was {minimum}.')
    return port
```

````python
#early return:조건이 부합하지 않으면 바로 return 을 하도록 하는 코딩 패턴
#이렇게 작성함으로써, 가독성이 올라감
def custom_sum(x,y):
    if (type(x) != int and type(x) != float) or (type(y) != int and type(y) != float):
        return 'error : 연산할 수 없습니다.'
    z = x + y 
    return z

print(custom_sum(5,'가나다'))
````

````python
#nonlocal
#inner가 outer에 변수를 건드리고 싶은데, 
#inner 안에서는 outer에 정의되어있는 x에 대한 수정권한이 없음
x = 100
def outer():
    x = 1
    def inner():
        nonlocal x#중첩함수 내에서 가장 가까운 상위함수의 변수에 접근이 가능함, 변경
        x += 1
        print(x)
    inner()
    print(x)

outer()
print(x)
````

````python
#global
x = 100
def outer():
    x = 1
    def inner():
        global x #global 변수에 접근이 가능해짐, 그리고 글로벌 변수 변경 가능
        x += 1
        print(x)
    inner()
    print(x)

outer()
print(x)
````

````python
#재귀의 비효율을 이렇게 해결할수 있다는 것을 보여주기 위함
#그러나 가능하면 for loop 이용해주세요 
from functools import lru_cache

# 데코레이터 사용함. 
@lru_cache(maxsize=None)#이 캐쉬를 이용해서 이미 계산된 것 memorise 후 꺼내쓰는것.
def fib(num):
    if num == 0:
        return 0
    if num == 1 or num == 2:
        return 1
    else:
        return fib(num-1) + fib(num-2)

fib(40)
````

## 깊은 복사와 얕은 복사

````python
#아무런 복사도 이뤄지지 않은 상태 , l & ll은 같은 곳을 가리킴
l = [[1,2,3],[4,5,6]]
ll = l

ll[0][0] =10
l, ll
````

````python
#얕은복사
l = [1,2,3,4,5,6]
ll = l.copy()

ll[0] =10
print(l,ll)
id(l[0]), id(ll[0]) #id값이 달라, 한 계층이니까 , 복사되서 다른계층을 가리킴
````

````python
#얕은복사
l = [[1,2,3],[4,5,6]]
ll = l.copy()


# ll[0] = 100 #첫번째 계층에서 변경, 복사가 됨. 그래서 하나 바꾸면 다른건 안바뀜
# ll[0][0] =10 #두번째 계층이니까, 첫번째 계층 이상의 깊이는 복사가 안됨, 그래서 하나바꾸면 다른것 변경됨.
print(l, ll)

print(id(l), id(ll)) #첫번째 계층, 다른주소, 그러므로 하나 바꾸면 다른건 안바뀜
id(l[0]), id(ll[0]) # 두번째 계층, 같은주소, 메모리 공유, 그러므로 하나 수정하면 다른것도 변경됨
````

````python
#얕은 복사는 한 뎁스만 카피, 1계층만 복사합니다.
#깊은 복사는 전체 뎁스 카피 , n 계층 모두 복사합니다.
````

````python
# 깊은 복사 deep copy
import copy

l = [[1, 2, 3], [4, 5, 6]]
ll = copy.deepcopy(l)
ll[0][0] = 10
print(l, ll)

print(id(l), id(ll)) #첫번째 다름 outer list level 첫번째 계층
print(id(l[0]), id(ll[0])) #두번째 다름 inner list level 두번째 계층
print(id(l[0][0]), id(ll[0][0])) #세번째 individual element level 
````

````python
# 얕은 복사
# 리스트 컴프리헨션은 안배운 내용이니 keep해두시고
# for문 배운다음 다시 오셔서 복습하시기를 권해드립니다.
l = [[1, 2, 3], [4, 5, 6]]
ll = [i for i in l]
l[0][0] = 100
print(l, ll)
````

````python
#위 list comprehension 을 아래와 같이 표현 가능, 어쨋든 이것도 shallow copy임.
l = [[1, 2, 3], [4, 5, 6]]
ll=[]
for i in l:
    ll += [i]
ll
````

````python
# 1단계 깊은 복사
# 리스트 컴프리헨션은 안배운 내용이니 keep해두시고
# for문 배운다음 다시 오셔서 복습하시기를 권해드립니다.
l = [[1, 2, 3], [4, 5, 6]]
ll = [i[:] for i in l] #리스트컴프리헨션을 만드는데, 그걸 다시 슬라이싱해서 안에꺼도 복사가 됨. 얕은복사 안 얕은복사가 이루어짐.
l[0][0] = 100
l, ll
````

## list comprehension

````python
#1 to 100까지 담겨있는 리스트 만들기

#1번: l = [1,2,3,4,,,100] #직접 손으로 , 노가다임..
#2번: list(range(1,101)) # 효율적임.
#3번: use loop #2번보단 비효율적
l=[]
for i in range(1,101):
    l.append(i)
l
````

- 참조값은 변경이 불가능(immutable) 합니다.
- 다른 자료형을 입력할 수 있으며, 튜플 안에 튜플로 다차원의 튜플을 만드는 것도 가능합니다.

````python
# 3.10버전 최신 문법에서는 match문법을 허용합니다!
def number_to_string(agrument):
    match agrument:
        case 0:
            return "zero"
        case 1:
            return "one"
        case 2:
            return "two"
        case default:
            return "nothing"

number_to_string(1)
number_to_string(5) #function안에 정의된 case가 아닌 것들을 입력하면 nothing 이라고 뜸.
````

## break 과 continue

````python
for i in range(10): #break 에 의해 for가 완료되지 못하고 중간에 interupted 되서 멈춤
    print(i)
    if i == 5:
        break
````

````python
#break에 의해 모든 for가 중단되는 것은 아님. 자신을 감싸고 있는 단일 반복문만 탈출함.(outer loop는 계속 진행)
for i in range(10):
    print("------")
    print(i)
    for j in range(10):
        print(i,j)
        if i ==5:
            break
````

````python
#continue는 다음loop으로 갑니다
#아래 코드에서 continue는 의미없습니다. 
#아래 print를 실행시키지 않고 다음 loop 으로 가게 됩니다. 그러므로 5는 print되지 않음
for i in range(10):
    if i ==5:
        continue
    print(i)
````

````python
# break => 반복문 중 break을 만나면 자신을 감싸는 반복문만 탈출하고, 밖의 반복문은 계속 실행됨
# continue => 반복문 중 만나면, 건너띄고 다음 loop으로 가게 됩니다.
````

````python
# 중복되는 문자열 중 가장 첫번째 나온
# 문자의 index를 반환하라!
s = 'abkkkbnbbbccdennnnneefefg'
ss = set(s)
while ss:
    string = ss.pop()
    print(string, s.index(string))
````

## args, kargs

* 가변 아규먼트, 가변 키워드 아규먼트

````python
def print_args(*args): #값들은 모두 tuple로 저장
    print(args)
    for x in args:
        print(x)

print_args(100, True, 'leehojun')
````

````python
def print_kwargs(a, **kwargs): #값들은 모두 dictionary로 저장 *1개만 사용하면 key만 넘어감
    print(a)
    print(kwargs)

print_kwargs(100, name='lee', age='10')
````

````python
#args 앞에 순서대로 넣을수 있는 변수가 와야 합니다. 순서가 바뀌면 에러
````

````python
#args 와 kwargs 둘다 사용하는 경우, 반드시 *args, **kwargs 순서대로 되어야 함. 반대는 에러남.
````

````python
def print_args_kwargs(*args, **kwargs):
    print('args:', args)
    for x in args:
        print(x)
    print('kwargs:', kwargs)
    for x in kwargs:
        print(x)
        print(kwargs[x])

print_args_kwargs(100, True, 'leehojun')
print('--------')
print_args_kwargs(score=100, name='leehojun', age='10')
print('--------')
print_args_kwargs(100, True, 'leehojun', score=100, name='leehojun', age='10')
````

````python
def print_args_kwargs(*args, **kwargs):
    print('args:', args)
    for x in args:
        print(x)
    print('kwargs:', kwargs)
    for x in kwargs:
        print(x)
        print(kwargs[x])
        
#아래와 같이 외부에 만들어서 넣어줘도 됨
inputlist=[100, True, 'leehojun']
inputdic={'score':100, 'name':'leehojun', 'age':'10'}
print_args_kwargs(*inputlist)
print('--------')
print_args_kwargs(**inputdic)
print('--------')
print_args_kwargs(*inputlist, **inputdic)
````

````python
# 광호님 코드
class BookInfo(object):
    def __init__(self, title, price, writer, publisher, count, soldout, buyer):
        self.title = title
        self.price = price
        self.writer = writer
        self.publisher = publisher
        self.count = count
        self.soldout = soldout
        self.buyer = buyer

    def __str__(self):
        return self.title

    def sell(self, buyer):
        if self.get_soldout():
            print('재고가 없습니다')
            return None
        
        self.count -= 1
        self.add_buyer(buyer)
        
        if self.count <= 0:
           self.set_soldout(True)

    def add_buyer(self, buyer):
        self.buyer.append(buyer)
    
    def set_soldout(self, value):
        self.soldout = value

    def get_soldout(self):
        return self.soldout
````

## iterator 

````python
#iterator : 값을 차례대로 꺼낼 수 있는 객체
#iterator를 만들기 위해 __iter__ ,  __next__ 2개의 method를 구현해야 함.

class MyIter:
    def __init__(self, stop):
        self.currentValue = 0
        self.stop = stop

    def __iter__(self):
        return self

    def __next__(self):
        if self.currentValue >= self.stop:
            raise StopIteration
        result = self.currentValue
        self.currentValue += 1
        return result

#다시 순회할 수 없음, 
li = MyIter(5)

for i in li:
    print(i)
````

````python
#다시 순회하려면 아래와 같이 init 함수 말고 iter 함수에 self.currntValue = 0으로 초기화해줘야함
class MyIter:
    def __init__(self, stop):
        self.stop = stop
    
    def __iter__(self):
        self.currentValue = 0
        return self

    def __next__(self):
        if self.currentValue >= self.stop:
            raise StopIteration
        result = self.currentValue
        self.currentValue += 1
        return result

li = MyIter(5)
# for는 iter먼저 실행하고, next로 StopIteration
for i in li:
    print(i)
for i in li:
    print(i)
````

## 제너레이터

````python
# yield를 만나면 카운트를 꺼낼 수 있음, 리턴값이 나오는거나 마찬가지임. 그러나 지금은 함수가 동작하고 있는 상태
# when creating generator use yield. when use yield, function can return intermediate result during run.
def gen():
    count = 0
    while True:
        yield count
        count+=1

gen()
````

````python
# 2중 decorator
def decorator1(func):
    def wrapper():
        print(f'deco1 > wrapper > func : {id(func)}')
        func() #이건 데코2의 wrapper 함수를 가리킴 
    print(f'deco1 > wrapper : {id(wrapper)}')
    return wrapper
 
def decorator2(func):
    def wrapper():
        print(f'deco2 > wrapper > func : {id(func)}')
        func()
    print(f'deco2 > wrapper : {id(wrapper)}')
    return wrapper
 
# 데코레이터를 여러 개 지정
@decorator1
@decorator2
def hello():
    print('hello')
 
hello()

# 헬로가 데코2 에 함수로 들어가서 데코2가 실행, 하여 데코2의 레퍼 리턴, 데코2의 래퍼가 데코원의 함수로 들어가 데코1 실행 
# 하여 데코1 의 래퍼가 리턴, 데코원의 래퍼가 실행, 데코1의 레퍼안의 함수는 데코2의 레퍼, 이게 실행되면 , 데코2의 래퍼안에 함수는 헬로 
# 다시 정리하면 데코1 래퍼 먼저 실행, 데코2 래퍼 두번째 실행, 헬로 실행
````

```python
@writefile
def add(a, b):
    return a + b

# writefile의 데코레이터 기능은 아래와 같은 형식으로 result.txt에 항상 저장되게 하는 것입니다.
# {
#     "a": 10,
#     "b": 20,
#     "a+b": 30
# }
```

````python
def writefile(func):
    def wrap_func(a,b):

        calc = func(a,b)
        #후처리를 하는 것  , data 1 , data2, data3
        #string
        string = ['a','b','a+b']
        #value
        value = [a,b,calc]
        s = dict(zip(string,value))
        #file creation
        f = open('result.txt', 'w')
        f.write('{')
        f.write('\n')
        for i,v in s.items():
            f.write(str(f'    "{i}":{v}'))
            f.write('\n')
        f.write('}')   
        f.close()

    return wrap_func


@writefile
def add(a, b):
    return a + b


add(10,20)
````
